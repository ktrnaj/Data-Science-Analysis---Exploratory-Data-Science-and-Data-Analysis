## Question: 1.What is the average salary across different experience levels and job titles, job categories over the past four years ?.

### CODE CELLS

```python
#Calculated average salary for each job title
AVG_SAL_JT = df.groupby('job_title')['salary'].mean().reset_index()

sdf = AVG_SAL_JT.sort_values(by='salary', ascending=False)
n = 10
top25JT = sdf.head(n)

OTH_JT = sdf[~sdf['job_title'].isin(top25JT['job_title'])]
OTH_avgSAL = OTH_JT['salary'].mean()

plt.figure(figsize=(12, 8))
plt.barh(top25JT['job_title'], top25JT['salary'], color='skyblue', label='Top 10')
plt.barh('Remaining Job Titles', OTH_avgSAL, color='lightgray', label='Remaining')
plt.xlabel('Average Salary', fontsize=14)
plt.ylabel('Job Title', fontsize=14)
plt.title('Average Salary by Job Title', fontsize=16)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()

#AVG_SAL_JT-> Average Salary by Job Title
#sdf-> Sorted df
#OTH_JT-> Other Job Titles 
#n -> Top 10
#OTH_avgSAL -> Average Salary by Other Job Title
#top25JT -> Top 25 Job Title
```

```python
#average salary for each Top job title along with experience level is calculated
AVGSAL_JT_EL = df.groupby(['job_title', 'experience_level'])['salary'].mean().reset_index()

TOP25jT = AVGSAL_JT_EL.nlargest(25, 'salary')  # Use 'salary' column for sorting

TOP25jT = TOP25jT.sort_values(by='salary', ascending=False)


pivot_df = TOP25jT.pivot(index='job_title', columns='experience_level', values='salary').fillna(0)
colors = ['#add8e6', '#6da1d9', '#3e75b0', '#003366']

# Plotted stacked bar chart with Experience level
plt.figure(figsize=(12, 8))
pivot_df.plot(kind='bar', stacked=True, color=colors, ax=plt.gca())
plt.xlabel('Job Title', fontsize=14)
plt.ylabel('Average Salary', fontsize=14)
plt.title('Average Salary by Top 25 Job Titles and Experience Level', fontsize=16)
plt.legend(title='Experience Level', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

#AVGSAL_JT_EL -> Average Salary by Job Title and Experience Level
#Top25jT -> Top 25 Job Title
#pivot_df-> pivoted Data frame
#Colors-> Chart colors
```

# QUESTION :Are there differences in Salaries and work settings based on demographic factors?

# CODE CELL

```python
# Count for employees who work in person and for employees who work remotely or have a different residence than the company location
in_p_EE = df[(df['work_setting'] == 'In-person') & (df['employee_residence'] == df['company_location'])]
in_p_EE_DL = df[(df['work_setting'] == 'In-person') & (df['employee_residence'] != df['company_location'])]

oth_EE = df[(df['work_setting'] != 'In-person') & (df['employee_residence'] == df['company_location'])]
oth_EE_DL = df[(df['work_setting'] != 'In-person') & (df['employee_residence'] != df['company_location'])]

#in_p_EE -> in person(worksetting) same residence and company location
#in_p_EE_DL ->in person(worksetting) same residence but different company location
#oth_EE -> Remote or Hybrid (worksetting) same residence and company location
#oth_EE_DL -> Remote or Hybrid (worksetting) same residence but different company location)
```

```python
# Count of employees work in person and have the same residence as the company location, 
#employees work in person but have a different residence than the company location,
#employees work remotely or have the same residence as the company location,
#employees work remotely and have a different residence than the company location are analysed

in_personsamelocct = in_p_EE.shape[0]
in_persondifflocct = in_p_EE_DL.shape[0]
other_samelocct = oth_EE.shape[0]
other_diffLocct = oth_EE_DL.shape[0]

print("Count of employees working in_person with the same residence location as the company location:", in_personsamelocct)
print("Count of employees working in_person with a different residence location than the company location:", in_persondifflocct)
print("Count of employees working remotely with the same residence location as the company location:", other_samelocct)
print("Count of employees working remotely with a different residence location than the company location:", other_diffLocct)

#in_personsamelocct ->in person(worksetting) same residence and company location
#in_persondifflocct ->in person(worksetting) same residence but different company location
#other_samelocct ->Remote or Hybrid (worksetting) same residence and company location
#other_diffLocct ->Remote or Hybrid (worksetting) same residence but different company location
```

```python
#Chart plotted for above data
labels = ['In person (Same residence Loc)', 'Remote (Same residence Loc)', 'Remote (Different residenceLoc)']
counts = [9406, 4673, 113]

plt.figure(figsize=(10, 6))
sns.barplot(x=labels, y=counts, hue=labels, dodge=False, palette='pastel', legend=False)
plt.title('Employee Work Settings and Residences')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

#labels -> labels for chart
#counts -> count of employees for chart
```

```python
#Average Salary for employees working in person with the same residence as the company location,
#for employees working in person with a different residence than the company location,
#for employees working remotely with the same residence as the company location
#employees working remotely with a different residence than the company location

in_person_sameres = df[(df['work_setting'] == 'In-person') & (df['employee_residence'] == df['company_location'])]

in_person_diffres = df[(df['work_setting'] == 'In-person') & (df['employee_residence'] != df['company_location'])]

rsameres = df[(df['work_setting'] != 'In-person') & (df['employee_residence'] == df['company_location'])]

rdiffres = df[(df['work_setting'] != 'In-person') & (df['employee_residence'] != df['company_location'])]


avgsal_in_person_sameres = in_person_sameres['salary'].mean()
avgsal_in_person_diffres = in_person_diffres['salary'].mean()
avgsal_remote_sameres = rsameres['salary'].mean()
avgsal_remote_diffres = rdiffres['salary'].mean()


print("Average Salary for Employees Working in-Person with the Same Residence as the Company Location:", avgsal_in_person_sameres)
print("Average Salary for Employees Working in-Person with a Different Residence than the Company Location:", avgsal_in_person_diffres)
print("Average Salary for Employees Working Remotely with the Same Residence as the Company Location:", avgsal_remote_sameres)
print("Average Salary for Employees Working Remotely with a Different Residence than the Company Location:", avgsal_remote_diffres)

#in_person_sameres ->in person(worksetting) same residence and company location
#in_person_diffres->in person(worksetting) same residence but different company location
#rsameres->Remote or Hybrid (worksetting) same residence and company location
#rdiffres->Remote or Hybrid (worksetting) same residence but different company location
#avgsal_in_person_sameres->in person(worksetting) same residence and company location
#avgsal_in_person_diffres->in person(worksetting) same residence but different company location
#avgsal_remote_sameres->Remote or Hybrid (worksetting) same residence and company location
#avgsal_remote_diffres->Remote or Hybrid (worksetting) same residence but different company location
```

```python
#Plotting Chart for above calculation
categs = ['In Person (Same Residence)', 'In Person (Different Residence)', 'Remote (Same Residence)', 'Remote (Different Residence)']
avgsal = [avgsal_in_person_sameres, avgsal_in_person_diffres,avgsal_remote_sameres, avgsal_remote_diffres]

pal = ['skyblue', 'lightgreen', 'lightcoral', 'lightsalmon']
plt.figure(figsize=(10, 6))
plt.pie(avgsal, labels=categs, autopct="%1.1f%%", startangle=140,colors=pal)

plt.title('Average Salary Distribution by Work Setting and Residence')

plt.axis('equal')  

plt.show()

#categs -> Lables for chart
#avgsal -> Counts for chart
#pal -> Colors for chart
```

# QUESTION :Are there regional disparities in experience level and job requirements?

# CODE CELL

```python
# Grouped the data by Top 5 company location, Experience level and number of jobs created by the country in each location and the average 
#salary for the same
locctz = df['company_location'].value_counts()

topl = locctz.nlargest(5).index

fig, axes = plt.subplots(nrows=len(topl), ncols=1, figsize=(10, 8), sharex=True)

for i, location in enumerate(topl):
    locd = df[df['company_location'] == location]
    expllsal = locd.groupby('experience_level')['salary'].agg(['mean', 'count'])

    ax = axes[i]
    ax.bar(expllsal.index, expllsal['mean'], color='skyblue', label='Average Salary')
    ax.set_ylabel('Average Salary')
    ax.set_title(f'Average Salary by Experience Level in {location}')

    ax_twin = ax.twinx()
    ax_twin.bar(expllsal.index, expllsal['count'], color='lightgreen', label='Job Count')
    ax_twin.set_ylabel('Job Count')

plt.tight_layout()
plt.show()

#locctz-location counts
#topl-Top locations
```

### EXPLANATION : The Table and chart shows the average salary and based on Experience level for United States, United Kingdom, Canada, Spain and Germany where over 80% jobs are created . However , there are high level of variations in the average salary for all the five countries . The United States has the highest average salary for all experience levels, followed by Canada, Germany, the United Kingdom, and Spain.Also The data is varied across experience levels and countries. For instance, the United States has the highest number of employees across all experience levels, while Spain and Germany have the least data points. 

# QUESTION :How do salary trends in USD reflect global market competitiveness?

# CODE 

```python
#Grouped data by company location,number of job titles and average salary (USD) for top 25 company location
grpdta = df.groupby('company_location')

loctd = grpdta.agg(
    nJT=('job_title', 'count'),
    avgsalusd=('salary_in_usd', 'mean')
)

top25ln = loctd.sort_values(by='avgsalusd', ascending=False).head(10)

coloc25 = top25ln.index.to_numpy()
nJT = top25ln['nJT'].to_numpy()
avgsalusd = top25ln['avgsalusd'].to_numpy()

plt.figure(figsize=(12, 6))

plt.plot(coloc25,avgsalusd, marker='o', label='Average Salary (USD)')
plt.ylabel('Average Salary (USD)')
plt.yscale('log') 

ax2 = plt.twinx()
ax2.bar(coloc25, nJT, color='skyblue', alpha=0.7, label='Number of Job Titles')
plt.ylabel('Number of Job Titles')

plt.xticks(rotation=45, ha='right')  
plt.tick_params(axis='x', which='major', labelrotation=45, labelsize=8)  


plt.title('Job Title Distribution and Average Salary by Company Location (Top 25)')
plt.legend()
plt.grid(True, linestyle='--', linewidth=0.5)
plt.tight_layout()

plt.show()

#grpdta ->Grouped Data
#loctd -> Location data
#nJT - Number of Job Title
#avgsalusd - Average Salary in USD
#top25ln - Top 25 company location
```

```python
#Grouped data by company location,number of job titles and average salary in OTHER CURRENCY for top 25 company location

gpdt = df.groupby('company_location')


locndta = gpdt.agg(
    nJtit=('job_title', 'count'),
    avgsalary=('salary', 'mean')  
)


top25locns = locndta.sort_values(by='avgsalary', ascending=False).head(10)


comlo = top25locns.index.to_numpy()
numjt = top25locns['nJtit'].to_numpy()
avgsly = top25locns['avgsalary'].to_numpy()


plt.figure(figsize=(12, 6))


plt.plot(comlo, avgsly, marker='o', label='Average Salary (Local Currency)') 
plt.ylabel('Average Salary (Local Currency)')


ax2 = plt.twinx()
ax2.bar(comlo, numjt, color='skyblue', alpha=0.7, label='Number of Job Titles')
plt.ylabel('Number of Job Titles')


plt.xticks(rotation=45, ha='right')  
plt.tick_params(axis='x', which='major', labelrotation=45, labelsize=8)  

plt.title('Job Title Distribution and Average Salary by Company Location (Top 25)')
plt.legend()
plt.grid(True, linestyle='--', linewidth=0.5)
plt.tight_layout()

plt.show()

#gpdt->Grouped Data
#locndta->location data
#nJtit->No of Job Title
#avgsly-> Average Salary
#avgsalary -> Average salary for top 25 locations
#comlo-> company location
#numjt-> No of Job Title
```
